```python
import pandas as pd
import seaborn as sns
df1 = pd.read_csv("Credit_card.csv")
df2= pd.read_csv("Credit_card_label.csv")
#merging two dataset in one using one common key 'Ind_ID'
df= pd.merge(df1,df2,on="Ind_ID")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# each record in these features are unique so drop them
df=df.drop(["Ind_ID","Mobile_phone",'Work_Phone',"Phone","EMAIL_ID"],axis=1)
df
```

```python
df["GENDER"].mode()
df["GENDER"]=df["GENDER"].fillna("F")
df["GENDER"].isnull().sum()
```

```python
# annual income feature have outlier so fill missing valu with median value
df['Annual_income']=df["Annual_income"].fillna(df["Annual_income"].median())
df["Annual_income"].isnull().sum()
```

```python
df["Type_Occupation"].mode()
df["Type_Occupation"]=df["Type_Occupation"].fillna("Laborers")
df["Type_Occupation"].isnull().sum()
```

```python
df["Birthday_count"]=df["Birthday_count"].fillna(df["Birthday_count"].mean())
df["Birthday_count"].isnull().sum()
```

```python
df.isnull().sum()
```

```python
#checking for outliers
sns.boxplot(x= df["Annual_income"])
```

```python
#treating outliers in Annual_salary
q1= df["Annual_income"].quantile(0.25)
q3= df["Annual_income"].quantile(0.75)
iqr= q3-q1
low_limit= q1-(1.5*iqr)
up_limit= q3+(1.5*iqr)
#capping method
df["Annual_income"]= df["Annual_income"].clip(lower=low_limit,upper=up_limit)
```

```python
#checking for outliers
sns.boxplot(x= df["Employed_days"])
```

```python
#treating outliers in employed_days
q1= df["Employed_days"].quantile(0.25)
q3= df["Employed_days"].quantile(0.75)
iqr= q3-q1
low_limit= q1-(1.5*iqr)
up_limit= q3+(1.5*iqr)
#capping method
df["Employed_days"]= df["Employed_days"].clip(lower=low_limit,upper=up_limit)
```

```python
#Data wrangling
# cheching for uniform distribution, checking skewness
df["Annual_income"].skew()
```

```python
df["Employed_days"].skew()
```

```python
df["Birthday_count"].skew()
```

```python
#treating right skewed feature (Annual_income)
import numpy as np
df["log_Annual_income"]=np.log(df["Annual_income"])
df["log_Annual_income"].skew()
```

```python
df["root_Annual_income"]=df["Annual_income"]**(1/2)
df["root_Annual_income"].skew()
```

```python
from scipy.stats import boxcox
df["box_cox_Annual_income"],param=boxcox(df["Annual_income"])
df["box_cox_Annual_income"].skew()
```

```python
#only keep boxcox drop others
df= df.drop(["log_Annual_income","root_Annual_income","Annual_income"],axis=1)
df
```

```python
from sklearn.feature_selection import chi2
f_p_values= chi2(x_train[["GENDER","Car_Owner","Propert_Owner","CHILDREN","Type_Income","Marital_status",
                          "Housing_type","EDUCATION","Family_Members","Type_Occupation"]],y_train)
f_p_values
# first array gives value of chi2, this should be high
# second array gives p value, it should be low
```

```python
#converting array to series
p_values =pd.Series(f_p_values[1])
gta=["GENDER","Car_Owner","Propert_Owner","CHILDREN","Type_Income",
                        "Marital_status","Housing_type","EDUCATION","Family_Members","Type_Occupation"]
p_values.index=x_train[gta].columns

p_values
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()


#scaling is only done on independent continuous variable
x_train[["Birthday_count","Employed_days",
         "box_cox_Annual_income"]]=sc.fit_transform(x_train[["Birthday_count","Employed_days","box_cox_Annual_income"]])
x_test[["Birthday_count","Employed_days",
        "box_cox_Annual_income"]]=sc.transform(x_test[["Birthday_count","Employed_days","box_cox_Annual_income"]])
```

```python
# 1. logistic regration
  #modeling
from sklearn.linear_model import LogisticRegression
l_reg= LogisticRegression()
l_reg.fit(x_train,y_train)
```

```python
  #prediction
l_reg_predTest= l_reg.predict(x_test)
l_reg_predTrain= l_reg.predict(x_train)
```

```python
#evaluation 
from sklearn.metrics import accuracy_score
print("Test Accuracy:",accuracy_score(y_test,l_reg_predTest))
print("Train Accuracy:",accuracy_score(y_train,l_reg_predTrain))

# to apply cv_score we need to combine test and train data set. and apply cv_score to original data set 
#  which is xx(independent variable dataset) and yy(dependent variable)
xx=pd.concat([x_train,x_test])
yy=pd.concat([y_train,y_test])

from sklearn.model_selection import cross_val_score
print("Cross validaton score:",cross_val_score(l_reg,xx,yy,cv=5,scoring="accuracy").mean())
```
